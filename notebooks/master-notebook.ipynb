# Caribbean Islands Climate EDA – Master Notebook
**Goal:** Understand temperature anomalies and precipitation patterns (1880–2024) across 15+ Caribbean islands/countries using NOAA/Global Historical Climatology Network data 
OR Model climate trends using real-world environmental data from NOAA, 
simulate future scenarios, compare popular time-series data machine learning models and create interactive geospatial visualisations. 
**Data source:** NOAA NCEI – Global Land Surface Temperature Anomalies + Precipitation -> [source](https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/global/time-series/caribbeanIslands/land_ocean/tavg/1/1/1850-2025)

## 1. Executive Summary – Key Climate Insights (updated as discovered)
- Caribbean has warmed **+1.4°C** since 1950 — faster than global land average   
- 2024-2025 was the hottest 24-month period on record for the whole Caribbean region
- Years have become hotter with less fluctuations like cool nights and winters.
